```python
#data from https://niagaraopendata.ca/dataset/2022-niagara-employment-inventory

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
```

```python
df=pd.read_csv('employment.csv')
```

```python
print(df.shape) #13033 rows and 19 
```

```python
#visualizing data
x=df['x'] #x column
y=df['y'] #y column

plt.scatter(x,y)
plt.xlabel("x column from csv")
plt.ylabel("y column from csv")
plt.title("Scatter plot of x and y")
plt.show()
```

```python
#histogram for primary sector 
plt.figure(figsize=(10,6))
df['primarysector'].value_counts().plot(kind='bar')
plt.title('Primary Sector')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['sizerangeemployees'].value_counts().plot(kind='bar')
plt.title('Distribution of Size Range Employees')
plt.xlabel('Size Range Employees')
plt.ylabel('Count')
plt.show()

df = df[df['sizerangeemployees'].str.isnumeric()]
```

```python
#data-group based on primary sector 
grouped_data=[df['sizerangeemployees'][df['primarysector']==sector]for sector in df['primarysector'].unique()]
#anova
result=f_oneway(*grouped_data)
#p-value for significance
if result.pvalue<0.5:
    print("statistically significant relationship")
else: 
    print("no statistically significant relationship")
```
